### Hướng dẫn sử dụng module `data_generating.py` (thuộc `/src/data_generating.py`)

```python
import sys
import os
sys.path.append(os.path.abspath("../src"))
from datetime import datetime
```

```python
from data_generating import generate_dataset
```

```python
df = generate_dataset(
    N=390,
    f10=100, mu1=-0.1, sigma1=0.015, alpha1=15,
    f20=150, mu2=-0.06, sigma2=0.04, alpha2=30,
    f30=200, mu3=-0.05, sigma3=0.01, alpha3=35,
    p_percent=0.1, L=200, I=5000,
    start_date="2024-01-01"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 7))  # Tăng chiều cao để hiển thị 3 biểu đồ rõ hơn

# Biểu đồ 1: Vàng
plt.subplot(2, 2, 1)
plt.plot(df['Date'], df['Vàng'], color='blue')
plt.title('Giá Vàng theo thời gian')
plt.xlabel('Ngày')
plt.ylabel('Giá (TMH/đơn vị)')
plt.grid(True)
plt.xticks(rotation=45)

# Biểu đồ 2: Tiền ảo
plt.subplot(2, 2, 2)
plt.plot(df['Date'], df['Tiền ảo'], color='green')
plt.title('Giá Tiền ảo theo thời gian')
plt.xlabel('Ngày')
plt.ylabel('Giá (TMH/đơn vị)')
plt.grid(True)
plt.xticks(rotation=45)

# Biểu đồ 3: Cổ phiếu
plt.subplot(2, 2, 3)
plt.plot(df['Date'], df['Cổ phiếu'], color='red')
plt.title('Giá Cổ phiếu theo thời gian')
plt.xlabel('Ngày')
plt.ylabel('Giá (TMH/đơn vị)')
plt.grid(True)
plt.xticks(rotation=45)

# Điều chỉnh layout để không bị chồng lấn
plt.tight_layout()
plt.show()
```

Cách xuất file dữ liệu

```python
version = datetime.now().strftime("%Y%m%d%H%M%S")
print(f"Version: {version}")
df.to_csv(f"../data/custom_dataset/scenario_dataset_{version}.csv", index=False)
```
