### Hướng dẫn sử dụng module `data_generating.py` (thuộc `/src/data_generating.py`)

```python
import sys
import os
sys.path.append(os.path.abspath("../src"))
from datetime import datetime
```

```python
from data_generating import generate_dataset
```

```python
df = generate_dataset(
    N=390,
    f10=100, mu1=0.2, sigma1=0.08/3, alpha1=5,
    f20=150, mu2=0.2, sigma2=0.18/3, alpha2=5,
    f30=200, mu3=0.15, sigma3=0.05/3, alpha3=5,
    p_percent=0.25, L=200, I=1000,
    start_date="2024-01-01"
)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 6))

plt.plot(df['Date'], df['Vàng'], label='Vàng', color='blue')
plt.plot(df['Date'], df['Tiền ảo'], label='Tiền ảo', color='green')
plt.plot(df['Date'], df['Cổ phiếu'], label='Cổ phiếu', color='red')

plt.title('Biến động giá của 3 loại tài sản theo thời gian')
plt.xlabel('Ngày')
plt.ylabel('Giá (TMH/đơn vị tài sản)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.xticks(rotation=45)  # Xoay trục x để dễ đọc ngày
plt.show()
```

Cách xuất file dữ liệu

```python
version = datetime.now().strftime("%Y%m%d%H%M%S")
print(f"Version: {version}")
df.to_csv(f"../data/custom_dataset/scenario_dataset_{version}.csv", index=False)
```
